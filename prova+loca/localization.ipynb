```python
class GradCAM:
  def __init__(self, model, target_layer):
    self.model = model
    self.target_layer = target_layer
    self.gradients = None
    self.activations = None

    #registra gli hook per intercettare i pesi intermedi della rete
    self.target_layer.register_forward_hook(self.save_activation)
    self.target_layer.register_full_backward_hook(self.save_gradient)

  #intercetta gli hook
  def save_activation(self, module, input, output):
    self.activations = output

  def save_gradient(self, module, grad_input, grad_output):
    self.gradients = grad_output[0]

  def generate_heatmap(self, input_image, target_class=None):
    self.model.eval()         #disattiva il dropout
    model_output = self.model(input_image)    #prende output

    if target_class is None:
      target_class = torch.argmax(model_output, dim=1).item()

    #calcola il gradiente per ottenere la variazione
    self.model.zero_grad()
    target = model_output[0][target_class]
    target.backward(retain_graph=True)
    #sposta i dati alla ram
    gradients = self.gradients.cpu().data.numpy()[0]
    activations = self.activations.cpu().data.numpy()[0]

    #calcola mean pooling per verificare se il canale è stato utile nella classificazione
    weights = np.mean(gradients, axis=(1, 2))

    #crea heatmap vuota
    heatmap = np.zeros(activations.shape[1:], dtype=np.float32)
    for i, w in enumerate(weights):
      heatmap += w * activations[i]
    #applica ReLU per scartare i valori fuori dai bound che non ci aiutano nella classificazione
    #heatmap = np.maximum(heatmap, 0)

    if np.max(heatmap) > 0:
      heatmap /= np.max(heatmap) # Normalizzazione
    return heatmap, target_class

#riporta l'immagine ai colori originali
def denormalize(img_tensor, mean, std):
  img = img_tensor.clone().squeeze().cpu().numpy().transpose(1, 2, 0)
  img = np.array(std) * img + np.array(mean)
  return np.clip(img, 0, 1)
```

```python
import os

target_layer = model[7][-1]
grad_cam = GradCAM(model, target_layer)    #crea l'oggetto sugli utlimi 4 livelli

#crea la cartella per salvare le immagini
dir = '../data/heatmapv3'
os.makedirs(dir, exist_ok=True)

cont = 1
idx_to_class = {v: k for k, v in test_dataset.class_to_idx.items()}

#scorre i batch
for images, labels in test_loader:

  #scorre le imamgini nei batch
  for idx in range(len(images)):
    true_label = labels[idx].item()
    class_name = idx_to_class[true_label]

    if 'good' in class_name.lower():   #crea heatmap solo per imamgini difettose
      continue

    img_tensor = images[idx].unsqueeze(0).to(device)
    #Grad-CAM
    heatmap, pred_class = grad_cam.generate_heatmap(img_tensor, target_class=None)

    img_rgb = denormalize(images[idx], imagenet_mean, imagenet_std)
    heatmap_resized = cv2.resize(heatmap, (img_rgb.shape[1], img_rgb.shape[0]))
    heatmap_uint8 = np.uint8(255 * heatmap_resized)

    heatmap_color = cv2.applyColorMap(heatmap_uint8, cv2.COLORMAP_JET)
    heatmap_color = np.float32(heatmap_color) / 255
    heatmap_color = heatmap_color[:, :, ::-1]

    overlayed_img = heatmap_color * 0.4 + img_rgb * 0.6
    overlayed_img = np.clip(overlayed_img, 0, 1)
    color_text = 'green' if true_label == pred_class else 'red'

    #crea grafico
    fig, axes = plt.subplots(1, 2, figsize=(8, 4))

    #stampa immagine
    axes[0].imshow(img_rgb)
    axes[0].set_title(f"Reale: {class_name}", color='black', fontweight='bold')
    axes[0].axis('off')

    #sovrappone heatmap
    axes[1].imshow(overlayed_img)
    axes[1].set_title(f"Predetto: {class_name[pred_class]}", color=color_text, fontweight='bold')
    axes[1].axis('off')
    plt.tight_layout()

    #salva immagine nella cartella
    nome_file = f"difetto_{cont}.png"
    percorso_completo = os.path.join(dir, nome_file)
    plt.savefig(percorso_completo, bbox_inches='tight', dpi=150)
    plt.close()
    cont += 1
```
